```python
# Importing All the Important Libraries 
import numpy as np
import pandas as pd
from time import sleep
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Key
# from selenium.webdriver.support.ui import WebDriverWait
# from selenium.webdriver.support import expected_conditions as EC
# from selenium.common.exceptions import TimeoutException
```

```python
Driver=webdriver.Chrome(options=webdriver.ChromeOptions())
Driver.get("https://www.swiggy.com")
sleep(3)
Driver.maximize_window()
Driver.find_element(By.XPATH,'//*[@id="location"]').send_keys('Bangalore')
sleep(5)
Driver.find_element(By.XPATH,'//*[@id="root"]/div[1]/div[1]/div/div[1]/div[1]/div/div[2]/div/div[2]/button[1]').click()
sleep(5)
html = Driver.find_element(By.TAG_NAME, 'html')
for i in range(80):
    html.send_keys(Keys.PAGE_DOWN)
    sleep(2)
sleep(10)
Urls = Driver.find_elements(By.CLASS_NAME,"_1j_Yo")
Detail=Driver.find_elements(By.CLASS_NAME,"_3Ztcd")
Rating= Driver.find_elements(By.CLASS_NAME,"_3Mn31")
print(Rating[0].text)
```

# CODE FOR TABLE 1

```python
Dlist=[]
for i in range(len(Detail)):
    url=Urls[i].get_attribute('href')
    item = Detail[i].text
    Name=item.split('\n')[0]
    Cousines=item.split('\n')[1]
    id=url.split('-')[-1]
    rat=Rating[i].text
    Star_rat=rat.split("\n")[0]
    Price_for_one=rat.split("\n")[-1]
    
    Dlist.append({"Restaurant_id" :id,
                  "Restaurant_link" :url,
                  "Retaurant Name":Name,
                  "Rating":Star_rat,
                  "price_for_one_two":Price_for_one,
                  "Cousines" :Cousines})
```

# TABLE 1 DATA

```python
df=pd.DataFrame(Dlist)
df.head(30)
```

```python
df.to_csv("Trial1.csv") ## SAVED CSV FILE OF TABLE 1 
```

# CODE FOR TABLE 2

```python
Elist=[]
for i in range(216,300):
    Driver=webdriver.Chrome(options=webdriver.ChromeOptions())
    Driver.get(df["Restaurant_link"].iloc[i])
    sleep(4)
    location=Driver.find_element(By.CLASS_NAME,"RestaurantNameAddress_area__2P9ib")
    Review_count=Driver.find_element(By.CLASS_NAME,"RestaurantRatings_totalRatings__3d6Zc")
   
   
    html = Driver.find_element(By.TAG_NAME, 'html')  
    for j in range(2):
        html.send_keys(Keys.PAGE_DOWN)
        sleep(2)
    dishes=Driver.find_elements(By.CLASS_NAME,"styles_itemNameText__3ZmZZ")
    prices=Driver.find_elements(By.CLASS_NAME, "rupee")
    for k in range(10):
        dish=dishes[k].text
        price=prices[k].text
#         print(df["Restaurant_link"].iloc[i])
#         print(dish)
#         print(price)
        
        Elist.append({"Restaurant_id":df["Restaurant_id"].iloc[i],
                     "Restaurant_Name":df["Retaurant Name"].iloc[i],
                     "location":location.text,
                     "dish_name":dish,
                     "Price":price,
                     "Review_count":Review_count.text})
        sleep(1)
    

    
   
    
    
    
    Driver.quit()
 
    
```
